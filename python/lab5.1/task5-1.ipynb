```python
import csv

with open('база_даних_книг.csv', 'w', newline='') as file:
    writer = csv.writer(file)
    writer.writerow(['Назва', 'Автор', 'Рік видання', 'Жанр'])

books = [
    ('Історія одного міста', 'Іван Іванов', 2005, 'Роман'),
    ('Кулінарний шедевр', 'Марія Петренко', 2012, 'Кулінарія'),
    ('Подорож до невідомого', 'Анна Сидоренко', 1999, 'Пригоди'),
    ('Психологія успіху', 'Олег Шевченко', 2018, 'Психологія'),
    ('Теорія космосу', 'Катерина Коваленко', 2008, 'Наукова'),
    ('Interstellar', 'Кіп Торн', 2015, 'Астрономія'),
    ('Ехопраксія', 'Пітер Уоттс', 2021, 'Наукова фантастика'),
    ('Пять ночей із Фредді', 'Скотт Коутон', 2022, 'Фентезі'),
    ('Блакитна цятка', 'Карл Саган', 2023, 'Астрономія'),
    ('Відьмак. Останнє бажання', 'Анджей Сапковський', 1993, 'Фентезі')
]

with open('база_даних_книг.csv', 'a', newline='') as file:
    writer = csv.writer(file)
    writer.writerows(books)

with open('база_даних_книг.csv', newline='') as file:
    reader = csv.reader(file)
    for row in reader:
        print(row)

def search_books(year):
    with open('база_даних_книг.csv', newline='') as file:
        reader = csv.reader(file)
        next(reader)  
        found_books = [row for row in reader if int(row[2]) == year]
    return found_books

found_books_1993 = search_books(1993)
print("Книги, видані у 1993 році:")
for book in found_books_1993:
    print(book)

def update_book_title(old_title, new_title):
    with open('база_даних_книг.csv', 'r', newline='') as file:
        reader = csv.reader(file)
        rows = list(reader)

    for row in rows:
        if row[0] == old_title:
            row[0] = new_title

    with open('база_даних_книг.csv', 'w', newline='') as file:
        writer = csv.writer(file)
        writer.writerows(rows)

update_book_title('Історія одного міста', 'Нова Історія')

def delete_book(title):
    with open('база_даних_книг.csv', 'r', newline='') as file:
        reader = csv.reader(file)
        rows = list(reader)

    for i, row in enumerate(rows):
        if row[0] == title:
            del rows[i]
            break

    with open('база_даних_книг.csv', 'w', newline='') as file:
        writer = csv.writer(file)
        writer.writerows(rows)

delete_book('Теорія космосу')

def sort_books_by_year():
    with open('база_даних_книг.csv', 'r', newline='') as file:
        reader = csv.reader(file)
        header = next(reader)
        rows = list(reader)

    rows.sort(key=lambda x: int(x[2]))  

    with open('база_даних_книг.csv', 'w', newline='') as file:
        writer = csv.writer(file)
        writer.writerow(header)
        writer.writerows(rows)

sort_books_by_year()

def count_books_by_genre():
    genre_count = {}
    with open('база_даних_книг.csv', newline='') as file:
        reader = csv.DictReader(file)
        for row in reader:
            genre = row['Жанр']
            if genre in genre_count:
                genre_count[genre] += 1
            else:
                genre_count[genre] = 1
    return genre_count

genre_statistics = count_books_by_genre()
print("Статистика книг у кожному жанрі:")
for genre, count in genre_statistics.items():
    print(f"Жанр: {genre}, Кількість книг: {count}")
```
